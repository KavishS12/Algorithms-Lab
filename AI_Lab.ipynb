```python
class graph:
    def __init__(self,vert):
        self.vert = vert
        self.adjList = [[] for _ in range(vert)]
        self.cost = {}
    def addEdge(self,u,v,cost = 0):
        self.adjList[u].append(v)
        self.cost[(u,v)] = cost
```

```python
#checking for cyclic grpah using dfs
cycle = False
def dfs(graph,node,visited):
    global cycle
    if node in visited:
        cycle = True
    if node not in visited:
        visited.append(node)
        for k in graph.adjList[node]:
                dfs(graph,k,visited)
    return visited

g = graph(4)
g.addEdge(0,1)
g.addEdge(1,2)
g.addEdge(2,0)
g.addEdge(0,2)
g.addEdge(2,3)
dfsRes = dfs(g,2,[])
print(dfsRes)
if(cycle is True):
    print("Graph has a cycle.")
else:
    print("No cycle.")
```

```python
def bfs(graph, start):
        frontier = [start]
        explored = set()
        explored.add(start)
        path=[]
        while frontier:
            current= frontier.pop(0)
            path = path + [current]
            for adj in graph.adjList[current]:
                if adj not in explored:                 
                    frontier.append((adj))
                    explored.add(adj)
        return path
    
g = graph(4)
g.addEdge(0,1)
g.addEdge(1,3)
g.addEdge(0,2)
bfsRes = bfs(g,0)
print(bfsRes)
```

```python
#uniform cost search
def ucs(graph,start,goal):
    visited = []
    queue = [(0,start,[start])]
    while queue:
        cost,node,path = queue.pop(0)
        visited.append(node)
        if node == goal:
            return path,cost
        for k in graph.adjList[node]:
            if k not in visited:
                queue.append((cost+graph.cost[node,k],k,path+[k]))
        queue.sort()
                
    return None,float('inf')

g = graph(7)
g.addEdge(0,1,17)
g.addEdge(1,0,15)
g.addEdge(2,0,53)
g.addEdge(2,1,40)
g.addEdge(3,1,46)
g.addEdge(4,3,3)
g.addEdge(2,4,31)
g.addEdge(4,5,29)
g.addEdge(6,4,8)
g.addEdge(3,6,11)
g.addEdge(5,6,40)
g.addEdge(5,2,17)

path,cost = ucs(g,6,0)
print(path)
print(cost)
```

```python
#water jug problem using bfs and dfs
class Jugs:
    def __init__(self,c1,c2,g1,g2):
        self.c1=c1
        self.c2=c2
        self.g1=g1
        self.g2=g2
        self.visited=set()
        self.queue=[((0,0),[(0,0)])]
        self.visited.add((0,0))
    
    def addstate(self,state):
        self.queue.append(state)
        self.visited.add(state[0])
    
    def isgoalstate(self,state):
        return state[0]==self.g1 and state[1]==self.g2
    
    def generateprods(self,state):
        prods=[]
        prods.append((self.c1,state[1]))
        prods.append((state[0],self.c2))
        prods.append((0,state[1]))
        prods.append((state[0],0))
        pour=min(state[0],self.c2-state[1])
        prods.append((state[0]-pour,state[1]+pour))
        pour=min(self.c1-state[0],state[1])
        prods.append((state[0]+pour,state[1]-pour))
        return prods
    
    def solveBFS(self):
        while self.queue:
            current,path=self.queue.pop(0)
            if self.isgoalstate(current):
                return path
            prods=self.generateprods(current)
            for prod in prods:
                if prod not in self.visited:
                    self.addstate((prod,path+[prod]))
        return None
    
    def solveDFS(self):
        while self.queue:
            current, path = self.queue.pop()  # pop the top element of the queue
            if self.isgoalstate(current):
                return path
            prods = self.generateprods(current)
            prods.reverse()  # Reverse the order to mimic DFS
            for prod in prods:
                if prod not in self.visited:
                    self.addstate((prod, path + [prod]))
        return None
    
j=Jugs(4,3,2,0)
#print(j.solveBFS())
print(j.solveDFS())
```

```python
# water jug problem using memoization
class Jugs:
    def __init__(self, c1, c2, g1, g2):
        self.c1 = c1
        self.c2 = c2
        self.g1 = g1
        self.g2 = g2
        self.memo = {}

    def solve(self):
        return self._solve(0, 0)

    def _solve(self, j1, j2):
        if (j1, j2) in self.memo:
            return self.memo[(j1, j2)]

        if j1 == self.g1 and j2 == self.g2:
            return [(j1, j2)]

        self.memo[(j1, j2)] = None

        for state in self.generate_states(j1, j2):
            path = self._solve(state[0], state[1])
            if path is not None:
                self.memo[(j1, j2)] = [(j1, j2)] + path
                return self.memo[(j1, j2)]

        return None

    def generate_states(self, j1, j2):
        states = []
        states.append((self.c1, j2))  # fill jug 1
        states.append((j1, self.c2))  # fill jug 2
        states.append((0, j2))  # empty jug 1
        states.append((j1, 0))  # empty jug 2
        pour = min(j1, self.c2 - j2)
        states.append((j1 - pour, j2 + pour))  # pour from jug 1 to jug 2
        pour = min(self.c1 - j1, j2)
        states.append((j1 + pour, j2 - pour))  # pour from jug 2 to jug 1
        return states

    def print_solution(self, solution):
        if solution is None:
            print("No solution exists.")
        else:
            print("Solution:")
            for i, state in enumerate(solution):
                print(f"{state[0]},{state[1]}")

jugs = Jugs(4, 3, 2, 0)
solution = jugs.solve()
jugs.print_solution(solution)
```

```python
#hill climb searching
def f(x):
    return x ** 2 - x ** 3

def hill_climb(f, low, high, d=0.02):
    x = 0
    while True:
        x1 = max((x - d, x, x + d), key=f)
        if x1 == x or x1 > high or x1 < low:
            return x
        x = x1

x_max = hill_climb(f, -10, 10)
print("max value", round(f(x_max), 2), "at x =", round(x_max, 2))
```

```python
import random

N = 8

def get_conflicts(state):
    conflicts = 0
    for i in range(N):
        for j in range(i+1, N):
            if state[i] == state[j] or abs(state[i] - state[j]) == abs(i-j):
                conflicts += 1
    return conflicts

def hill_climbing_search():
    state = [random.randint(0, N-1) for _ in range(N)]
    
    current_state = state
    best_state = state
    
    current_conflicts = get_conflicts(current_state)
    best_conflicts = current_conflicts
    
    iter = 0
    while best_conflicts != 0 and iter<1000:
        neighbors = []
        for i in range(N):
            for j in range(N):
                if j != current_state[i]:
                    neighbor = current_state[:]
                    neighbor[i] = j
                    neighbors.append(neighbor)
        
        best_neighbor = None
        best_neighbor_conflicts = float('inf')
        for neighbor in neighbors:
            conflicts = get_conflicts(neighbor)
            if conflicts < best_neighbor_conflicts:
                best_neighbor = neighbor
                best_neighbor_conflicts = conflicts
        
        if best_neighbor_conflicts < current_conflicts:
            current_state = best_neighbor
            current_conflicts = best_neighbor_conflicts
            if best_neighbor_conflicts < best_conflicts:
                best_state = best_neighbor
                best_conflicts = best_neighbor_conflicts
        else:
            current_state = [random.randint(0, N-1) for _ in range(N)]
            current_conflicts = get_conflicts(current_state)
        iter += 1
    
    if best_conflicts == 0:
        print("Solution found:")
        for i in range(N):
            row = ['#'] * N
            row[best_state[i]] = 'Q'
            print(''.join(row))
    else:
        print("No solution.")
    print("Iterations = " ,iter)

hill_climbing_search()
```

```python
def is_valid(board):
    n = len(board)
    for i in range(n):
        for j in range(i+1,n):
            if board[i] == board[j] or abs(board[i]-board[j])==abs(i-j):
                return False
    return True

def solveNQueens(n):
    count = 0
    solns = []
    queue = [[]]
    
    while queue:
        board = queue.pop(0)
        row = len(board)
        if row == n:
            solns.append(board)
            count += 1
        else:
            for col in range(n):
                if is_valid(board + [col]):
                    queue.append(board + [col])
    
    return solns, count

def printResult(solns, count):
    for sol in solns:
        print('Solution:', tuple(sol))
    print('Number of solutions found:', count)

# Example usage:
n = 8
solns, count = solveNQueens(n)
printResult(solns, count)
```
